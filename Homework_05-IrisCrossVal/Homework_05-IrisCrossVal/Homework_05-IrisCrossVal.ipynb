# import modules

```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

# helper function(s)

# read iris data into Pandas data frame

```python
df = pd.read_csv('iris.data',
names=['sepal length','sepal width','petal length','petal width','class'],
index_col=None)
```

# plot data to investigate pattern(s)

# split data into feature matrix (X) and target labels (y)

```python
Y = df['class']
X = df.drop('class',axis=1)
mapping = {'Iris-setosa':'r', 'Iris-versicolor':'g','Iris-virginica':'b'}
color = [mapping[s] for s in Y]
pd.plotting.scatter_matrix(df, color=color)
```

# create decision tree classifier

```python
model = tree.DecisionTreeClassifier()
```

# use K-fold cross validation to test classifier

```python
yp = cross_val_predict(model, X, Y,cv=10)
```

# print results: accuracy, precision, and recall 

```python
acc = accuracy_score(Y, yp)
pre = precision_score(Y, yp, average='macro')
rec = recall_score(Y, yp, average='macro')
print('accuracy = %.2f, precision = %.2f, recall = %.2f' % (acc, pre, rec))
```
